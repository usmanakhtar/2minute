```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.rcParams["figure.figsize"] = [12.50, 3.50]
plt.rcParams["figure.autolayout"] = True

solo = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data/KPI-6-WaitingTime/Waiting_Solo-Rides.csv')

profitmax =pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data/KPI-6-WaitingTime/Waiting_Profit-Maximization.csv')
pooled =pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data/KPI-6-WaitingTime/Waiting_Pooled-Rides.csv')

for col in solo.columns:
    solo[col]=solo[col]//3600
    solo.head()

for col in profitmax.columns:
    profitmax[col]=profitmax[col]//3600
    profitmax.head()


for col in pooled.columns:
    pooled[col]=pooled[col]//3600
    pooled.head()

fig, (ax1, ax2, ax3) = plt.subplots(ncols=3)
fig.subplots_adjust(wspace=1.01)


s = sns.heatmap(profitmax, annot=True,  fmt=".2f", cmap='viridis', ax=ax1)
s.set(xlabel='number of Travellers', ylabel='number of Drivers', title = 'Waiting time of Profit Maximization')

s = sns.heatmap(pooled, annot=True, fmt=".2f", cmap='viridis', ax=ax2)
s.set(xlabel='number of Travellers', ylabel='number of Drivers', title = 'Waiting time of   Pooled Rides')

s = sns.heatmap(solo, annot=True, fmt=".2f", cmap='viridis', ax=ax3)
s.set(xlabel='number of Travellers', ylabel='number of Drivers', title = 'Waiting time of Solo Rides')

txt="I need the caption to be present a little below X-axis"
#ax3.yaxis.tick_right()

fig.subplots_adjust(wspace=0.001)
plt.show()
#fig.savefig('ServiceRate.png', dpi=600)
```
